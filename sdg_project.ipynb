```python
# 🔰 SDG 13: Simple CO2 Emissions Predictor
# Goal: Use ML to predict CO2 emissions based on energy use

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 📊 1. Sample dataset
data = {
    'Energy_Use_per_Capita': [3000, 4000, 5000, 6000, 7000],
    'CO2_Emissions_per_Capita': [5.2, 6.5, 7.0, 8.4, 9.0]
}
df = pd.DataFrame(data)

# 🧹 2. Prepare data
X = df[['Energy_Use_per_Capita']]
y = df['CO2_Emissions_per_Capita']

# 📦 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# 🔍 4. Train model
model = LinearRegression()
model.fit(X_train, y_train)

# 📈 5. Predict & Visualize
predictions = model.predict(X_test)

plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, predictions, color='red', linewidth=2, label='Predicted')
plt.xlabel('Energy Use per Capita')
plt.ylabel('CO₂ Emissions per Capita')
plt.title('CO₂ Emissions Prediction')
plt.legend()
plt.grid(True)
plt.show()

# 🧠 6. Print score
print("Model Accuracy (R²):", model.score(X_test, y_test))
```
